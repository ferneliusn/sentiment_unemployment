# Get List Of Historic URLs to Scrape From

```python
years_wanted = ['2016', '2015', '2014','2013','2012','2011','2010','2009','2008',
               '2007','2006','2005','2004','2003','2002','2001',
                '1999', '1998','1997','1996','1995','1994', '1993'
               ]
```

```python
def get_urls(years_wanted):
    hist_urls = []
    for year in years_wanted:
        historic_doc = requests.get('https://www.federalreserve.gov/monetarypolicy/fomchistorical' + year + '.htm').text
        historic_soup = BeautifulSoup(historic_doc, 'html.parser')
        for link in historic_soup.find_all('a'):
            hist_urls.append(link.get('href'))
    hist_urls = [i for i in hist_urls if i is not None]
    final_hist = [word for word in hist_urls if 'fomcminutes'in word]
    final_hist = [link for link in final_hist if link.endswith('.htm')]
    return final_hist
    
```

```python
final_hist = get_urls(years_wanted)
```

```python
print(final_hist)
```

```python
def get_urls2(years_wanted):
    hist_urls = []
    for year in years_wanted:
        historic_doc = requests.get('https://www.federalreserve.gov/monetarypolicy/fomchistorical' + year + '.htm').text
        historic_soup = BeautifulSoup(historic_doc, 'html.parser')
        for link in historic_soup.find_all('a'):
            hist_urls.append(link.get('href'))
    hist_urls = [i for i in hist_urls if i is not None]
    final_hist = [word for word in hist_urls if '/fomc/minutes/'in word]
    final_hist = [link for link in final_hist if link.endswith('.htm')]
    return final_hist
    
```

```python
to_add = get_urls2(years_wanted)
```

# Make final list of URLs

```python
final_urls = [*final, *final_hist, *to_add] 
```

```python
print(final_urls)
```

# Scrape and Save Text To Text Files

```python
def scrape(urls):
    for url in urls:
        url = 'https://www.federalreserve.gov' + url
        html_doc = requests.get(url).text
        soup = BeautifulSoup(html_doc, 'html.parser')
        text = soup.get_text()
        name = url[-12:-4]
        text_file = open(name + ".txt", "w")
        n = text_file.write(text)
        text_file.close()                   
```

```python
scrape(final_urls)
```
