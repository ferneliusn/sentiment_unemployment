# 5 Unit Horizon

```python
train_data_5 = train_data
```

```python
train_data_5["UNRATE-1"] = train_data.UNRATE.shift(1)
train_data_5["UNRATE-3"] = train_data.UNRATE.shift(3)
train_data_5["UNRATE-6"] = train_data.UNRATE.shift(6)
train_data_5["FEDFUNDS-1"] = train_data.FEDFUNDS.shift(1)
train_data_5["FEDFUNDS-3"] = train_data.FEDFUNDS.shift(3)
train_data_5["FEDFUNDS-6"] = train_data.FEDFUNDS.shift(6)
train_data_5["FEDFUNDS-10"] = train_data.FEDFUNDS.shift(10)
train_data_5["future_rate"] = train_data.UNRATE.shift(-5)
train_data_5["past_rate"] = train_data.UNRATE.shift(10)
train_data_5=train_data_5.dropna(axis=0)
```

```python
train_data_5
```

```python
train_data_5['sasdate'] = pd.to_datetime(train_data_5['sasdate']).astype(int)/ 10**9
```

```python
train_data_5.to_csv('./CSVs/train_data_5.csv')
```
