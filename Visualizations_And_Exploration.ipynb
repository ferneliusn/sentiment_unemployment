```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv('./CSVs/train_data_plot.csv')
train_data = train_data.drop(['Unnamed: 0'], axis=1)
plotting_data = train_data
y_train = train_data['future_rate']
```

```python
train_data
```

```python
train_data.sort_values(by='negative_sentiment', ascending=True)
```

```python
train_data.sort_values(by='negative_sentiment', ascending=False)
```

## Correlation Scatterplots

```python
train_data['cur-past'] =train_data['unemployment_rate']- train_data['past_rate']
train_data['fut-cur'] = y_train - train_data['unemployment_rate']
```

```python
plt.style.use('seaborn-darkgrid')
plt.scatter(x=train_data['negative_sentiment'],y=train_data['cur-past'])
plt.xlabel('Negative Sentiment Score')
plt.ylabel('Difference')
plt.title('Negative Sentiment Score vs. Difference Between Current Rate and Rate 10 Statements Ago')
```

```python
plt.style.use('seaborn-darkgrid')
plt.scatter(x=train_data['negative_sentiment'],y=train_data['fut-cur'])
plt.xlabel('Negative Sentiment Score')
plt.ylabel('Difference')
plt.title('Negative Sentiment Score vs. Difference Between Rate 10 Statements From Now and the Current Rate')
```

```python
plt.style.use('seaborn-darkgrid')
plt.scatter(x=train_data['negative_sentiment'],y=train_data['FEDFUNDS'])
plt.xlabel('Negative Sentiment Score')
plt.ylabel('Federal Funds Rate')
plt.title('Negative Sentiment Score vs. Federal Funds Rate')
```

```python
plt.style.use('seaborn-darkgrid')
plt.scatter(train_data['unemployment_rate'],train_data['negative_sentiment'])
plt.xlabel('Unemployment Rate')
plt.ylabel('Negative Sentiment Score')
plt.title('Unemployment Rate vs. Negative Sentiment Score')
```

## Time Series Comparisons
